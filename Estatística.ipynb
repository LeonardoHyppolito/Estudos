*  Objetivo da estatística
1. Fornecer Métodos, para lidar de modo racional, com a variabilidade. 
2. Descobrindo regularidade, isto é, padrões. 
3. Expressar matematicamente, padrões de comportamento (regularidade), que estão presentes nos dados, mas que não são facilmente, identificáveis. 

   



*  Conceitos Básicos
1. População: Conjunto de todos os elementos envolvidos, com pelo menos uma característica em comum, com a possibilidade de ser finita ou infinita. Exemplo: Em uma pesquisa eleitoral na cidade de são paulo, a população seria TODOS os eleitores da cidade de SP, a característica seria votar em SP, e o eleitor é visto como uma unidade. 
2. Amostra: Uma parte da população que representa o TODO., para isso a amostra deve ser convenientemente escolhida, e apresentar a mesma característica da população. 
3. Como obter uma boa amostra: 
Metodologia: Amostragem; existem parâmetros que conseguimos obter a partir da população ( média, variância, desvio padrão), e também podemos obter algumas estimativas de parâmetros, que são calculados em cima da amostra. 
4. Parâmetros: Constantes que caracterizam uma população, em geral são valores desconhecidos no início do projeto, e representamos eles através de letras gregas. 
5. Estimativas: Bom, como a amostra é uma representação de uma população, as estimativas são os valores calculados em cima da amostra, isto é, a ideia é fornecer informações sobre os parâmetros. Vale notar, que como as estimativas são calculadas em cima da amostra, estas estão sujeitas a erros. Geralmente, esses erros estão relacionados a erros de amostragem, isto é, obter uma amostra que não representa, ou representa mal, uma população. 

*  Divisão da estatística ( Olhar Superficial).
1. Estatística Descritiva: Essa parte da estatística, basicamente trabalha com a apresentação de tabelas e gráficos. Resumo, ou descrição dos dados através de medidas descritivas( que serão abordadas ao longo do curso). E em geral, como o nome sugere, a ideia principal é descrever, isto é, não tem por objetivo tirar conclusões, seja de uma população, ou de um subconjunto desta. 
2. Inferência Estatística: Já essa parte da estatística tem como ideia geral, fornecer métodos que nos permitam realizar inferências sobre a população, a partir de amostras. Em suma, utilizar os métodos, principalmente a probabilidade, em cima das amostras, para tirar conclusões sobre a população. 
 2.1: Estimação dos parâmetros, isto é, estimar os parâmetros de uma população, através das amostras. 
 2.2: Testes de Hipóteses: Basicamente, vão nos possibilitar testar os métodos que vamos utilizar, para obter respostas. Além de tentar entender o que a população em relação aquela amostra, nos fornece de informação.

*  Características e Variáveis 

1. Característica: É uma propriedade qualquer das unidades de uma população, porém o interesse não é nas unidades em si, mas sim nas suas características. 
  1.1 As características das unidades são muito vastas, por exemplo, sexo masculino, região onde mora, escolaridade, estado civil etc. Porém, nem todas são relevantes para a pesquisa em si, logo temos de selecionar as mais relevantes. 
2. Variável: As possibilidades para uma determinada característica, isto é, se em uma pesquisa, é relevante perguntarmos às unidades, a região onde ela mora, as possibilidades de resposta (variáveis) seriam: os bairros, as ruas, a região. 




*  Escalas de medidas das características 

1. Escala Nominal: Números funcionam como rótulo, ou como números mesmo. Por 



exemplo, podemos rotular o sexo masculino sendo 2 e o feminino como 1. {Feminino: 1 , Masculino : 2}. Sendo assim, não podemos ordenar estes números, justamente por serem apenas uma forma de identificação, da mesma maneira não é possível realizar cáculos com eles. Servem apenas para contagem. 
2. Escala Ordinal: Os números, além de rótulos, permitem uma ordenação, isto é, é possível associar a eles, valores que que representam uma ordem. Exemplo:"excelente" = 1, "bom" = 2, "ruim" = 3, note que esses números servem como rótulo, mas tambem "pertencem" a uma determinada ordenação. As operações aritméticas não fazem sentido, isto é,se somarmos um livro bom com um livro ruim, isso não resultará em um livro mediano, não faz sentido. 
3. Escala Intervalar: Além de ordenar as unidades quanto a característica, essas possuem unidades de medida. O zero, nesse tipo de escala é apenas uma convenção.Por exemplo: quando tratamos de temperatura, C° ou F°, o valor zero difere-se de uma para a outra. Outro exemplo bom, é a contagem, de tempo, diversas sociedades com o passar do tempo, utilizaram diversas formas de contar o tempo, isto é o primeiro dia do ano, por exemplo é diferente aqui no Brasil e na China. A escala intervalar, nos permite calcular as diferenças entre os dados, por exemplo: qual a diferença entre 30 C° e 30 F° ?. Não possui um zero natural. 

4. Escala de racional (Razão), escala mais abrangente de todas: Ordena as unidades quanto à característica mensurada, possui uma unidade de medida (mensuração), e sua origem é única. Exemplo: Peso, minutos, km... Inclui a escala intervalar, porém a escala racional possui um zero verdadeiro, que neste caso significa a ausência de quantidade. Tanto as diferenças, tanto as razões tem significado, fazem sentido. *exemplo: No exemplo, dos eleitores de são paulo, as unidades de observação, seriam os eleitores de São paulo. *

3. Conjunto de dados: Conjunto de todas obsrvações. 
*exemplo: seguindo o exemplo dado a cima, vimos que as observações são o conjunto de respostas de um determiado indivíduo. Logo, o conjunto de dados são o conjunto de respostas de todos os indivíduos.*

*   DISTRIBUIÇÃO DE FREQUÊNCIAS 



1. Forma de resumir uma informação sobre uma ou mais variáveis, organizando o conjunto de dados em classes, indicando a frequência da observação em cada classe. 

2. O objetivo da distribuição de frequência é representar a forma como os dados (as variáveis) se distribuem. A partir dessa representação, nó conseguimos entender qual a melhor forma, o melhor modelo de distribuição de probabilidade mais adequado para os dados que estamos utilizando. Qual técnica utilizar, como extrair informação. 

3. Geralmente, existem dois tipos de tabelas de distribuição de freqências: A simples e a Cruzada. 



*   Tabela de Classificação simples: Variáveis Categóricas. 


1. Para montar uma tabela de classificação simples ( com uma variável categórica), basta considerar j = número como rótulo para as classes disponíveis (coluna 2). No exemplo dado em aula, as possíveis classes eram: Bom, Ruim, Ótimo. Logo a primeira coluna da tabela, tinha o título J, onde a primeira linha era o número  1, a segunda linha o número  2  e a terceira linha o número 3. A segunda coluna da tabela, apresentava as possíveis classes, e a terceira coluna,  a respectiva frequência de cada uma das classes, isto é, basta contar o número de vezes que cada classe foi observada. A frequencia absoluta é representada por Fj. 



2. Podemos então alimentar a tabela supracitada, criando novas colunas, a primeira que é interessante acrescentar é a *frequencia acumulada*, isto é, conforme vamos descendo a coluna, vamos somando os valores das frequências, e o somatório final, será obviamente, o número de observações total. 



3. Após a coluna de frequência acumulada, podemos criar uma coluna de Frequência relativa, que indica a relação do número de observações de determinada classe, com o número de observações total (%). 


4.  E por fim, podemos adiconar mais uma coluna a tabela, com o nome de frequência relativa acumulada, isto é, vamos somando as proporções das observações (%) de cada classe, até que no fim das observações, obviamente, um total de 1 ou 100. 








*   TABELAS DE DISTRIBUIÇÃO DE FREQUÊNCIAS PARA VARIÁVEIS DISCRETAS. 

1. Os passos para a construção de uma tabela de frequência para variáveis discretas, são os mesmos do que para as variáveis categóricas, a única diferenã, é que vamos rotular as classes com números, e não mais como "Bom, Ruim, otimo" como visto anteriormente. 


2. A partir daí basta contar os elementos pertencentes a cada classe, e segui o passo a passo: contar quantas observações pertencem a determinada classe, colocar isso na frequencia absoluta, ir somando as frequencias absolutas para montar a coluna da frequência acumulada; calcular a proporção da frequencia absoluta observada em cada classe com o total de observação (proporção), para montar a coluna das frequência relativas , e somar as respectivas proporções para obter a coluna da tabela relativa acumulada. 





*  TABELA DE DITRIBUIÇÃO DE FREQUÊNCIA PARA VARIÁVEIS CONTÍNUAS.

Para facilitar a visualização é interessante ordenar o conjunto de observações em ordem crescente, para facilitar essa vizualização.

*   MEDIDAS DESCRITIVAS 



1. A estatística descritiva, em geral, busca sintetizar dados brutos. "Elegante em sua simplicidade, mas limitada naquilo que pode nos dizer" - Charles Wheelan. 

2. Cálculos tecnicamente bem feitos podem não responder as nossas perguntas. 

3. Quando pegamos um problema de análise de dados, pode ser uma boa escolha iniciar o projeto buscando as medidas de tendência central, que intuitivamente, ou não, achamos mais relevantes para responder a questão inicial. 

4. Média aritmética: Simples, porém propensa a distorções causadas pelos outliers. 

5. A mediana divide um conjunto de dados ordenados ao meio, e isso faz com que ela não seja afetada por distorções. ( Essa é a grande vantagem da mediana). 

6. As estatísticas descritivas nos dão a percepção de alguns fenômenos, mas não nos diz tudo sobre eles, são boa para efeito comparativo. 

7. Do ponto de vista da Acurácia, a questão media x mediana, gira em torno de se os outliers numa distribuição distorcem o que está sendo descrito, ou se são, ao contrário parte importante da mensagem. 

8. Um questionamento interessante que aparece no livro "Estatística, o que é, para que serve, e como funciona" - Charles Wheelan, sobre a estatística descritiva é " A simplicidade e a facilidade de uso introduzidas ao reduzir muitos indicadores em um único numero, superam a inerente inacurácia no processo ?". 

9. A moda é o valor que mais aparece em um conjunto de dados. Quando estamos tratando de variáveis contínuas, é comum que agrupemos os dados em classes, e se vizualizarmos a distribuição de frequência, o retângulo com maior altura, representa a classe modal.

*  MEDIDAS DE DISPERSÃO


1. Amplitude Total: fornece uma ideia rudimentar, de variação. 
  1.1 A amplitude total é uma medida pouco precisa, e muito sensível a valores discrepantes.
  1.2 A única ideia que vejo de interessante, é que ela nos diz qual a mair diferença existente entre os dados. 
  1.3 Para calcular, basta ordenar os dados, e fazer Maior numero - Menor numero. 

2. Amplitude Interquartílica (aq): Diferença entre o terceiro e o primeiro quartil, isto é  aq = q3 - q1. 
  2.1 50% dos valores mais centrais dos valores diferem no máximo em X, onde X = aq 
  2.2 a grande desvantagem dessa amplitude está nas situações onde temos dados muito próximos ao mínimo e o máximo, isto é, essa medida não os leva em consideração. 
  2.3 Ja a vantagem, é justamente não ser sensíveis aos valores discrepantes. 

3. Variância:
