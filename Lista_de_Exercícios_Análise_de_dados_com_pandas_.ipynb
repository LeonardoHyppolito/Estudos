Exercício 18: Verifique se todas as observações na coluna "cidade" são iguais a "São Paulo". Salve a reposta em uma variável.

```python
sp = df.cidade == "São Paulo"
print(sp) # Porém desse jeito, estamos vendo linha a linha, quais são True e quais são false. 
```

```python
# Se quisermos uma variavel, que armazene um True caso todas as observações de cidade == Sao Paulo, e false caso uma 
# Delas não seja, então temos de usar o all()
sp_corrigido = (df.cidade == "São Paulo").all()
print(sp_corrigido)
```
