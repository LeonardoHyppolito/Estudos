```python
# Primeiro vamos chamar a biblioteca pandas para trabalhar com a gente. 
import pandas as pd
```

```python
# Agora, nos definimos uma variável "df", sendo o que o pandas lê do arquivo csv que importamos. 
df = pd.read_csv("campeonato-brasileiro-full.csv")
```

```python
# Método head.(), por padrão nos mostra as 5 primeiras linhas do df, ou se quisermos passamos o numero de linhas desejado como parâmetro.
df.head()
```

```python
#Método .tail(), nos mostra as ultimas linhas do df, a quantidade padrão é 5, ou o numero de linhas desejado é passado como parãmetro. 
df.tail()
```

insight: Observando o conjunto de dados comparando o resultado do .head() com o resultado do .tail() podemos pensar que possivelmente temos mais dados faltantes nos primeiros anos do que nos anos mais recentes. 

**Primeira Inspeção nos dados**

```python
# Este método nos permite com apenas uma linha de código, obter indormações gerais do nosso conjunto de dados. Um bom começo ! 
df.info()
```

insight: Aqui descobrimos que o df possui 18 colunas, 4 delas com valores do tipo inteiro, e 14 delas com valores do tipo objeto. 
  Podemos levantar um questionamento: **Por que algumas colunas, por exemplo, as de index: 7, 8, 9, 10 possuem menos linhas ?**

```python
df.describe()
```

```python
# Por padrão o df.describe() nos retorna algumas estatísticas descritivas, porém somente das variáveis numéricas. 
```

```python
#Existe um parâmetro que podemos passar, para que o df.describe() também analise as features de valores não numericos. 
```

```python
# Aqui passamos um parâmetro "include" = O, pois como vimos no df.info as demais features possuem valores do tipo Objeto. 
df.describe(include="O")
```

```python
# Podemos também passar um "all", como parâmetro do include, e isso fará com que o describe retorne todo o conjunto de dados. 
df.describe(include="all")
```

```python
# Porém é importante reparar que muitos dados estão com o valor "Nan", isso porque muitas dessas variáveis não possuem estatísticas 
# Por exemplo: Não faz sentido pensar na media dos mandantes. 
```



*   Existem duas formas de acessar colunas específicas: 


1.   Conceito de colchetes e aspas. 
Este modo é interessante para casos em que as colunas do df não seguem as normas da linguagem python, isto é, por exemplo, colunas com nomes composto que são separadas por espaços e não por "_". Neste caso, a única forma de acessa-las é com o conceito de colchetes e aspas, como se estivéssemos acessando um pedaço do df. 





```python
# Utilzando colchetes 
df["rodada"]
```

```python
df["visitante"]
```

```python
# Utilizando o conceito de ponto "."
df.visitante
```

```python
# Agora repare que se pegarmos umaa suposta coluna " mandante placar", este método não funcionaria. Isso retornará um erro !
df.mandante placar
```

```python
# Nesse caso as únicas maneiras de acessar acessar a coluna seria com colchetes e aspas ou renomeando, e substituindo "  " por " _ "
```

```python
# selecionando mais de uma coluna: Basta adicionar um colchetes, e separa-las por virgula.
df[["mandante", "rodada", "visitante"]]
```
