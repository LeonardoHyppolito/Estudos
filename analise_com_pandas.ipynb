**FILTRANDO DATA FRAME - UMA CONDIÇÃO**

```python
df = pd.read_csv("campeonato-brasileiro-full.csv")
```

```python
df.head()
```

```python
#Uma maneira de filtrar, é utilizando df.NomeDaColuna, o computador percorre a coluna dia, e retorna true ou false com base no == domingo. 
df.dia == "Domingo"
```

```python
#Porém, as vezes pode ser interessante visualizar todas as informações que estão relacionadas com o dia Domingo: 
df[df.dia =="Domingo"]
```

```python
#Também podemos fazer isso, utilizando o .loc 
df.loc[df.visitante == "Sao Paulo"]
```

```python
# Outra possibilidade interessante,  é que podemos filtrar mais de uma observação específica: 
df.loc[df.visitante == "Sao Paulo" , "formacao_visitante"]
```

```python
# Também podemos criar uma variável que receba uma indormação sobre uma determinada coluna, e nos retorne true ou false. 
sp_visita = df.visitante == "Sao Paulo"  # Veirifica quando o são paulo é visitante, e armazena-o. 
```

```python
sp_visita  # Variável que armazena todos as observações que contém o sao paulo visitante. 
```

```python
# Agora podemos criar um DataFrame, somente com as observações que contém o São paulo de visitante. 
```

```python
df_sp_visita = df.loc[sp_visita]   # isso porque armazenamos na variável sp_visita, como sendo todas as informações que ontém na coluna visitante 
# O valor São paulo.  # "crie um DataFrame", pegando toda a informação dentro do Dataframe original, que contenha "sp_visita"
```

```python
# Aqui o Dataframe, que conntém todos os jogos do Sao Paulo, do dataframe original, em que o São paulo é o visitante. 
df_sp_visita
```

```python
#Por curiosidade, eu quero saber todas as vezes que o São Paulo utilizou a formação 4-4-2, em um jogo como visitante
formacao4_4_2 = df_sp_visita.loc[df_sp_visita.formacao_visitante == "4-4-2"]
```

```python
# Armazenei isso em um variável, e se eu printar a variável que é um Df, teremos todos os jogos em que o São paulo foi visitante, e utilizou a formação 
formacao4_4_2
```
