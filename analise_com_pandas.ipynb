**Filtrando DataFrame, múltiplas condições**

```python
df = pd.read_csv("campeonato-brasileiro-full.csv")
```

```python
df.head()
```

```python
# Variável, vá ao meu df.coluna mandante, e analisa pra mim, os valores Sao Paulo. 
mandante_sp = df.mandante == "Sao Paulo"
```

```python
# Como atribuímos a essa variável todas as ocasiões, em que o valor Sao Paulo aparece na feature "mandante", ela funciona como um booliano, 
#Como se ela percorresse toda a coluna, verificando se existe (TRUE) ou não existe (FALSE) o valor Sao Paulo, naquela linha. Logo se dermos um .head():
mandante_sp.head() 
```

```python
# Agora como a ideia é analisar, múltiplas condições, podemos criar uma nova variável, com uma nova condição. A ideia
# é verificar se o time fez pelo menos 2 gols, 
gols_feitos = df.mandante_placar >= 2 
```

```python
# Da mesma forma, a variável percorre a coluna, verificando se o valor é >= 2 e nos retorna TRUE ou FALSE. 
gols_feitos.head()
```

```python
# Aqui o pc esta percorrendo as colunas de interesse, e verificando se satisfaz ambas condições, detalhe para o "&"
# que funcionou e não consegui aplicar o "and" o qual eu achava que era a mesma coisa.
(mandante_sp & gols_feitos).head()
```

```python
# Podemos então criar uma nova variável, chamada sp gol feitos, juntando essas duas.
# Essa variável pede ao Df todas as observações que satisfazem ambas as condições. 
sp_gols_feitos = df.loc[(mandante_sp & gols_feitos)]
```

```python
sp_gols_feitos
```

```python
# Podemos também, pedir ao df, somente algumas colunas das observações que satisfazem ambas condições, por exemplo: 
sp_gols_feitos = df.loc[(mandante_sp & gols_feitos),["hora", "dia", "visitante"]]
```

```python
#Aqui, vamos observar apenas a hora, o dia e o visitante, que sofreu pelo menos 2 gols do Sao Paulo, atuando como mandante. 
sp_gols_feitos
```

```python
# A partir disso, temos um pedaço interessante do Df, onde podemos aplicar métodos e ir investigando os casos. 
sp_gols_feitos.info()
#Então, sabemos que foram 185 vezes que o Sao Paulo fez mais de 2 gols como mandante; e que os valores das colunas escolhidas são do tipo object
```

```python
# Devido as colunas escolhidas, esse .describe não faz muito sentido, uma vez que os valores são do tipo object. 
sp_gols_feitos.describe()
# Mas em outras ocasiões, como o caso de colunas do tipo int, pode ser bastante útil ter esse resumo
```
