1) Calcule para todas as observações a diferença entre as colunas Mandante Placar e Visitante Placar. Faça isso preenchendo as lacunas!

```python
df.loc[:, ["mandante_placar","visitante_placar"]] # Utilizando o loc[], para selecionar as colunas desejadas através do nome. 
# O primeiro parâmetro nos diz quantas informações eu quero pegar, e quais são elas, passando : , (vazio dos dois lados), ele pegará todas as 
#informações das colunas escolhidas. 
```

```python
#Também poderiamos fazer utilizando o iloc[], e passando os indices das colunas desejadas. 
```

```python
# Fazendo a função de subtração, de modo anonima. (lambda)
df.loc[:, ["mandante_placar","visitante_placar"]].apply(lambda x: x.max() - x.min(), axis=1) # Ja que queremos a diferença em módulo. 
# O computador nos retorna o index e o resultado da subtração do maximo e do mínimo. axis=1, pois precisamos dos valores das linhas. 
```
