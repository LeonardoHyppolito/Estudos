1) Calcule para todas as observações a diferença entre as colunas Mandante Placar e Visitante Placar. Faça isso preenchendo as lacunas!

6) Selecione todos os jogos com "Estado Mandante" sendo o RJ que ocorreram no Sábado

```python
# Para isso podemos criar variáveis que armazenam as condições e depois unir com o conceito de and e or, nesse caso invés de and, utilizamos "&"
```

```python
estado_mandante = (df.mandante_estado == "RJ")
dia_jogo = (df.dia == "Sabado")
```

```python
# Conferindo se alguma das informações das primeiras linhas, satisfazem as duas condições. 
(estado_mandante & dia_jogo).head() # Se uma das condições for falsa, ele nos retorna false. 
```

```python
# Agora podemos criar um df, somente com as informações que satisafazem as duas condições utilizando o loc[]
df_rj_sabado = df.loc[estado_mandante & dia_jogo]
```

```python
df_rj_sabado.head(3)
```

```python
# Outra forma de realizar isso é voltando a colocar o mandante_estado e o dia como index. 
df.set_index(["dia", "mandante_estado"], inplace = True)
```

```python
df.head(3)
```

```python
# Para facilitar, podemos utilizar um swaplevel, para inverter a ordem dos index. 
df = df.swaplevel()
```

```python
df.head() # Visualizando o swap. 
```

```python
# Agora podemos fazer um df.loc[], e acessar rapidamente todos os jogos com RJ de mandante. 
df.loc["RJ"] # Note que ele não nos retorna o index RJ, isso porque, todas as informações são referentes a mandante_estado = RJ. 
```

```python
# Com isso fica mais fácil de acessar múltiplas informações, ao invés de fazer tudo o que foi feito anteriormente.  
df.loc[("RJ", "Sabado")] # Muito mais simples, e de fácil visualização. Logo, entendimento mais claro das informações. 
```

9) Converta todos os elementos na coluna Estado Mandante para lower!

```python
# Para acessar o estado mandante como coluna precisamos resetar o nosso index. 
df.reset_index(inplace=True)
```

```python
df.head(3) # Agora nem o dia nem o mandante estado estarão como index. 
```

```python
# Agora basta acessar a coluna mandante_estado e aplicar o metodo lower, basicamente transformar os elementos da coluna em minusculas. 
df.mandante_estado = df.mandante_estado.str.lower() # Precisamos transformar em str, para aplicar o método lower. 
```

```python
df.head(3) # Agora os valores da coluna mandante_estado estarão em letras minúsculas. 
```

10) Verifique quantos dias na coluna "Dia" contêm "feira"!

```python
# Python nos oferece o método contains, que verifica se "contém" algo em determinada coluna, neste caso, podemos pedir para o python para ele verificar
# Se "contém" ou melhor contains alguma str "feira" e pedir para ele somar quantas observações contains "feira" utilizando o sum()
df.dia.str.contains("feira").sum() # Importante notar que contains é aplicada somente em str, então precisamos transformar antes de chamar a função. 
```
